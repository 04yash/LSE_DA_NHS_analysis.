```python
# Import necessary libraries.
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Set figure size.
sns.set(rc={'figure.figsize':(15, 13)})

# Set the plot style as white.
sns.set_style('white')
```

```python
# Import the necessary packages
import pandas as pd

# Import the movies data set
nc = pd.read_excel('national_categories.xlsx')

# View the DataFrame
nc.head()
print(nc.columns)
```

# simple lineplot along context type and appointments.

```python
# Create a simple lineplot.
sns.lineplot(x='context_type', y='appointment_month', data=nc, ci=None)
```

# # Plot the appointments over the available date range, and review the context types for months.

```python
# Create lineplots with specification.
sns.lineplot(x = 'appointment_month', y = 'count_of_appointments',
             data=nc[nc['context_type'].isin(['Care Related Encounter', 'Inconsistent Mapping','Unmapped'])],
             hue ='context_type', ci=None)
sns.lineplot(x = 'appointment_month', y = 'count_of_appointments',
             data=nc[nc['context_type'].isin(['Care Related Encounter', 'Inconsistent Mapping','Unmapped'])],
             hue ='context_type',style = 'context_type', markers=True, ci=0)
```

# Plot the appointments over the available date range, and review the service settings for months.

```python
# Create lineplots with specification.
sns.lineplot(x = 'appointment_month', y = 'count_of_appointments',
             data=nc[nc['service_setting'].isin(['General Practice', 'Extended Access Provision','Other','Unmapped','Primary Care Network'])],
             hue ='service_setting', ci=None)
sns.lineplot(x = 'appointment_month', y = 'count_of_appointments',
             data=nc[nc['service_setting'].isin(['General Practice', 'Extended Access Provision','Other','Unmapped','Primary Care Network'])],
             hue ='service_setting',style = 'service_setting', markers=True, ci=0)
```

# linePlot of appointments over the available date range, and reviewing the national categories for months.

```python
# Set figure size.
sns.set(rc={'figure.figsize':(20, 18)})

# Set the plot style as white.
sns.set_style('white')
# Create lineplots with specification.
sns.lineplot(x = 'appointment_month', y = 'count_of_appointments',
             data=nc[nc['national_category'].isin(['Care Home Needs Assessment & Personalised Care and Support Planning', 'Care Home Visit','Clinical Triage','General Consultation Acute','General Consultation Routine','Group Consultation and Group Education','Home Visit','Inconsistent Mapping','Non-contractual chargeable work','Patient contact during Care Home Round','Planned Clinical Procedure','Planned Clinics','Service provided by organisation external to the practice','Social Prescribing Service','Structured Medication Review','Unmapped','Unplanned Clinical Activity','Walk-in'])],
             hue ='national_category', ci=None)
sns.lineplot(x = 'appointment_month', y = 'count_of_appointments',
             data=nc[nc['national_category'].isin(['Care Home Needs Assessment & Personalised Care and Support Planning', 'Care Home Visit','Clinical Triage','General Consultation Acute','General Consultation Routine','Group Consultation and Group Education','Home Visit','Inconsistent Mapping','Non-contractual chargeable work','Patient contact during Care Home Round','Planned Clinical Procedure','Planned Clinics','Service provided by organisation external to the practice','Social Prescribing Service','Structured Medication Review','Unmapped','Unplanned Clinical Activity','Walk-in'])],
             hue ='national_category',style = 'national_category', markers=True, ci=0)
```
