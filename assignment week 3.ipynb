```python
# Import Pandas.
import pandas as pd
import numpy as np
# Import the movies data set
nc1 = pd.read_excel('national_categories.xlsx')

# Read CSV file from the current working directory.
ad1 = pd.read_csv('actual_duration.csv')

# Read CSV file from the current working directory.
ar1 = pd.read_csv('appointments_regional.csv')



# Data imported correctly?
print(nc1.columns)
print(nc1.dtypes)
print(nc1.shape)
print(ad1.columns)
print(ad1.dtypes)
print(ad1.shape)
print(ar1.columns)
print(ar1.dtypes)
print(ar1.shape)
```

#  Between what dates were appointments scheduled?

```python
# Determining the minimum and maximum dates in the ad DataFrame.

# max = Returns max of a column
print(ad1['appointment_date'].max())

# min = Returns min of a column
print(ad1['appointment_date'].min())


```

# grouping the dates with max and min count of appointments on the day in the actual duration dataframe.


```python
# Determining the minimum and maximum dates in the ad DataFrame.
# grouping the dates with max and min count of appointments on the day.



ad1.groupby('appointment_date').count_of_appointments.agg(
    dates_max=('max'),
    dates_min=('min'),
)
```

```python
# Determining the minimum and maximum dates in the nc DataFrame.

# max = Returns max of a column
print(nc1['appointment_date'].max())

# min = Returns min of a column
print(nc1['appointment_date'].min())

```

# grouping the dates with max and min count of appointments on the day in the national categories(nc1) dataframe

```python
# Determining the minimum and maximum dates in the nc DataFrame.
# grouping the dates with max and min count of appointments on the day.



nc1.groupby('appointment_date').count_of_appointments.agg(
    dates_max=('max'),
    dates_min=('min'),
)
```

# Which service setting was the most popular for NHS North West London from 1 January to 1 June 2022?

```python
# Concatenate the three DataFrames.
nc2_ad2_ar2 = pd.concat([nc1, ad1,ar1], axis=0)

# View the new DataFrame.
print(nc2_ad2_ar2.shape)
nc2_ad2_ar2.fillna(0)
```

```python
# Convert the data type of column 'Date' from string (YYYY/MM/DD) to datetime64
nc2_ad2_ar2["appointment_month"] =  pd.to_datetime(nc2_ad2_ar2["appointment_month"], format="%Y/%m/%d")
nc2_ad2_ar2.dtypes
```

```python
# Create group 1.
group1 = nc2_ad2_ar2.groupby(['service_setting', 'count_of_appointments','appointment_month'])

# View the DataFrame.
group1
group1.sum()
```

# Which month had the highest number of appointments?

# the month of october 2021-10-01 had the highest number of count of appointments with 4387 unmapped services in the NHS North West London area.

# What was the total number of records per month?

```python
# Create group 1.
group1 = nc2_ad2_ar2.groupby(['service_setting', 'count_of_appointments','appointment_month',]) \
.agg(['sum', 'mean']) \
.sort_values('appointment_month',ascending=False)
group1.fillna(0)
```

# the total record per month for general practice recorded in june 2022 was 5048
# the total record per month for primary care network in june 2022 was 647
